```python
#lineplot
import matplotlib.pyplot as plt
x=[98,67,88,95,88]
x1=[59,88,87,97,55]
y=["part1","part2","part3","part4","part5"]

plt.plot(y , x,marker="^",label="min")
plt.plot(y , x1,marker="^",label="max")
plt.legend()
plt.xlabel("parts of harry potter")
plt.ylabel("populatity")
plt.title("Harry Potter")
plt.show()
```
